```python
from pathlib import Path
def get_latest_log():
    return max(filter(lambda p: p.is_file() and p.name.startswith("simulation_results") and p.suffix == ".pickle", Path.cwd().iterdir()), key=lambda p: p.lstat().st_mtime)
```

```python
import pickle
import matplotlib.pyplot as plt
p = get_latest_log()
print(p)
with open(p, mode='br') as f:
    data = pickle.load(f)
        
```

```python
sim = data[0]
```

```python
fig, ax = plt.subplots()
X = [float(x) for x in sim["stats"]['acceptance_probability'] if float(x) != 1.0]
ax.plot(X)
```

```python
fig, ax = plt.subplots()
n = 100
w = len(sim["stats"]['acceptance_probability']) // n
X1 = [len([x for x in sim["stats"]['acceptance_probability'][i*n:i*n+w] if float(x) == 1.0]) for i in range(n)]
X2 = [len([x for x in sim["stats"]['acceptance_probability'][i*n:i*n+w] if float(x) != 1.0]) for i in range(n)]
ax.plot(X1)
ax.plot(X2)
```

```python
fig, ax = plt.subplots()
X = [float(x) for x in sim["stats"]['new_cost']]
ax.plot(X)
```

```python
fig, ax = plt.subplots()
X = [float(x) for x in sim["stats"]['curr_cost']]
maxx = max(X)
X = [x/maxx for x in X]
ax.plot(X)
```
