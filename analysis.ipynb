```python
import json
num_datasets = 8
data = [None]*num_datasets
for i in range(num_datasets):
    with open(f'stats{i}.json', 'r') as f:
        data[i] = json.load(f)
```

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots()
for i in range(num_datasets):
    X = [float(x) for x in data[i]['acceptance_probability'] if float(x) != 1.0]
    ax.plot(X)
```

```python
fig, ax = plt.subplots()
for k in range(num_datasets):
    n = 100
    w = len(data[k]['acceptance_probability']) // n
    X1 = [len([x for x in data[k]['acceptance_probability'][i*n:i*n+w] if float(x) == 1.0]) for i in range(100)]
    X2 = [len([x for x in data[k]['acceptance_probability'][i*n:i*n+w] if float(x) != 1.0]) for i in range(100)]
    ax.plot(X1)
    ax.plot(X2)
```

```python
fig, ax = plt.subplots()
for k in range(num_datasets):
    X = [float(x) for x in data[k]['new_cost']]
    ax.plot(X)
```

```python
fig, ax = plt.subplots()
for k in range(num_datasets):
    X = [float(x) for x in data[k]['curr_cost']]
    maxx = max(X)
    X = [x/maxx for x in X]
    ax.plot(X)
```
