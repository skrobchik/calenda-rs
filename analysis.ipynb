```python
import json
with open('stats.json', 'r') as f:
    data = json.load(f)
```

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots()
X = [float(x) for x in data['acceptance_probability'] if float(x) != 1.0]
ax.plot(X)
```

```python
fig, ax = plt.subplots()
n = 100
w = len(data['acceptance_probability']) // n
X1 = [len([x for x in data['acceptance_probability'][i*n:i*n+w] if float(x) == 1.0]) for i in range(100)]
X2 = [len([x for x in data['acceptance_probability'][i*n:i*n+w] if float(x) != 1.0]) for i in range(100)]
ax.plot(X1)
ax.plot(X2)
```
