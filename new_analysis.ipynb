```python
import json
import matplotlib.pyplot as plt
with open('results3.json') as f:
    data = json.load(f)
```

```python
len(data)
```

```python
fig, ax = plt.subplots(dpi=300)
for i in range(len(data)-10, len(data)):
    final_cost = data[i]["best_schedule_cost"]
    steps = data[i]["best_schedule_run_report"]["num_steps"]
    curr_cost = data[i]["best_schedule_run_report"]["stats"]["curr_cost"]
    ax.plot(curr_cost, label=f'{steps} steps (final_cost = {final_cost})')
ax.legend(fontsize=5)
ax.set_title('curr_cost')
ax.set_xlabel('x')
ax.set_ylabel('cost')
# ax.set_yscale('log')
```

```python
fig, ax = plt.subplots()
x = [data[i]["best_schedule_run_report"]["num_steps"] for i in range(len(data)-10, len(data))]
y = [data[i]["best_schedule_cost"] for i in range(len(data)-10, len(data))]
ax.plot(x, y)
ax.set_xlabel('Num. Simulation Steps')
ax.set_ylabel('Final Cost')
ax.set_title('Simulation steps vs Final Cost')
# ax.set_yscale('log')
# ax.set_xscale('log')
```

```python
fig, ax = plt.subplots()
final_cost = data[0]["best_schedule_cost"]
steps = data[0]["best_schedule_run_report"]["num_steps"]
curr_cost = data[0]["best_schedule_run_report"]["stats"]["curr_cost"]
temperature = data[0]["best_schedule_run_report"]["stats"]["temperature"]
ax.plot(temperature, label=f'{steps} steps (final_cost = {final_cost})')
ax.legend(fontsize=5)
ax.set_title('curr_cost')
ax.set_xlabel('x')
ax.set_ylabel('cost')
```
