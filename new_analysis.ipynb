```python
import json
import matplotlib.pyplot as plt
with open('results_1.json') as f:
    data = json.load(f)
```

```python
len(data)
```

```python
data[0]['simulation_options']['total_steps']
```

```python
fig, ax = plt.subplots(dpi=300)
for i in range(len(data)-10, len(data)):
    final_cost = data[i]['final_cost']
    steps = data[i]['simulation_options']['total_steps']
    curr_cost = data[i]['stats']['curr_cost']
    ax.plot(curr_cost, label=f'{steps} steps (final_cost = {final_cost})')
ax.legend(fontsize=5)
ax.set_title('curr_cost')
ax.set_xlabel('x')
ax.set_ylabel('cost')
ax.set_yscale('log')
```

```python
fig, ax = plt.subplots()
x = [data[i]['simulation_options']['total_steps'] for i in range(len(data)-10, len(data))]
y = [data[i]['final_cost'] for i in range(len(data)-10, len(data))]
ax.plot(x, y)
ax.set_xlabel('Num. Simulation Steps')
ax.set_ylabel('Final Cost')
ax.set_title('Simulation steps vs Final Cost')
# ax.set_yscale('log')
# ax.set_xscale('log')
```

```python
fig, ax = plt.subplots()
final_cost = data[-1]['final_cost']
steps = data[-1]['simulation_options']['total_steps']
curr_cost = data[-1]['stats']['curr_cost']
temperature = data[-1]['stats']['temperature']
ax.plot(temperature, label=f'{steps} steps (final_cost = {final_cost})')
ax.legend()
ax.set_title('curr_cost')
ax.set_xlabel('x')
ax.set_ylabel('cost')
```
